# Relationship between kvalue and number of instances. 

```python
 
# creating dataframe
df_1 = pd.DataFrame({
    'Datasets': ['Breast', 'Diabetes', 'Fruit', 'Heart', 'Iris','stroke'],
    
    'NoOfInstances': [569,768,59,303,150,5110],
    'kvalue' : [5,17,3,5,11,23]
    
})
df_1copy=df_1.copy(deep=True)
df_1['NoOfInstances'] = np.log2(df_1['NoOfInstances'])


df_1['kvalue'] = np.log2(df_1['kvalue'])

  
# plotting graph
df_1.plot(x="Datasets", y=[ "NoOfInstances","kvalue"], kind="bar")
```

```python
#3Scatter plot
df_1.plot.scatter(x='NoOfInstances', y= 'kvalue')
```

```python
# Relationship between No. of instances and optimum K using T-test
instance_mean=df_1copy['NoOfInstances'].mean()
k_mean=df_1copy['kvalue'].mean()

print ('The mean of instances {} and k value is {}'.format(instance_mean, k_mean))
```

```python
Hypothesis= '''
The Null hypotheiss is H0 : There is no linear correlation between Number of Instances and optimum k
The alternative hypotheiss is H1: There is linear correlation between No of Instances and optimum k
'''
print (Hypothesis)
```

```python
ttest,pval = ttest_rel(df_1copy['NoOfInstances'],df_1copy['kvalue'])
print("t-value",ttest)
print("p-value",pval)

if pval <0.01:
  print("we reject null hypothesis (and accept alternative hypothesis")
else:
  print("we accept null hypothesis")
```

# Relationship between k value and number of features using T-test. 

```python

  
# creating dataframe using number of features and optimum k for each dataset
df_2 = pd.DataFrame({
    'Datasets': ['Breast', 'Diabetes', 'Fruit', 'Heart', 'Iris','stroke'],
    
    'Features': [32,9,7,14,5,12],
    'kvalue':[5,17,3,5,11,23]
})
  

# plotting graph
df_2.plot(x="Datasets", y=[ "Features","kvalue"], kind="bar")
```

```python
# Relationship between No. of Features affect the Optimum K using T-test 
Hypothesis= '''
The Null hypotheiss is H0 : There is No linear correlation between Number of features and optimum k
The alternative hypotheiss is H1: There is linear correlation between No of features and optimum k
'''
print (Hypothesis)
```

```python
feature_mean=df_2['Features'].mean()
k_mean=df_2['kvalue'].mean()

print ('The mean of instances {} and k value is {}'.format(feature_mean, k_mean))
```

```python
ttest,pval = ttest_rel(df_2['Features'],df_2['kvalue'])
print("t-value",ttest)
print("p-value",pval)

if pval <0.01:
  print("we reject null hypothesis (and accept alternative hypothesis")
else:
  print("we accept null hypothesis")
```
